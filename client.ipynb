```python
import requests
import base64

import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
data = {'img': base64.b64encode(requests.get('http://services.fms.gov.ru/services/captcha.jpg').content).decode('utf-8')}

img_b64 = data['img']
im_bytes = base64.b64decode(img_b64)
im_arr = np.frombuffer(im_bytes, dtype=np.uint8)
image = cv2.imdecode(im_arr, flags=cv2.IMREAD_GRAYSCALE)
plt.imshow(image, cmap='gray')
plt.show()

response = requests.post('http://127.0.0.1:5000/api/captcha/solve', json = data)
response.text
```
